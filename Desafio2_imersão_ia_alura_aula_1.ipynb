Para melhor visualização, separarei cada parte do codigo.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lendo o arquivo CSV
data = pd.read_csv("carro.csv", sep=";")
```

```python
# 1. Carro mais vendido (assumindo que a quantidade está relacionada à aparição na lista)
mais_vendido = data["Marca"].value_counts().index[0]
print(f"Carro mais vendido: {mais_vendido}")
```

```python
# 2. Gasto total por carro (agrupado)
gastos_por_carro = data.groupby("Marca")["Preço"].sum().sort_values(ascending=False)
gastos_por_carro = gastos_por_carro.to_frame().reset_index()
print(gastos_por_carro)
```

```python
# 3. Gráfico de barras horizontais
plt.figure(figsize=(10, 5))
plt.barh(gastos_por_carro["Marca"], gastos_por_carro["Preço"])
plt.xlabel("Gasto Total")
plt.ylabel("Marca do Carro")
plt.title("Gasto Total por Marca de Carro")
plt.show()
```

```python
# Salvando a nova tabela em CSV
gastos_por_carro.to_csv("gastos_por_carro.csv", sep=";", index=False)
```
